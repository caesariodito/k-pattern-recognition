# functions lama (euclidean, similaritas, dkk)

```python
# Perhitungan Euclidean
def euclidean(x1,x2):
  return np.sqrt(np.sum((x1-x2)**2))

# Menghitung seluruh Euclidean Daun A,B,C,D
def count_all_euclidean(input,db):
  out = [] # array kosong
  for i in range(input.shape[0]):
    out.append([]) # menambahkan panjang array dimensi pertama
    for j in range(input.shape[1]):
      out[i].append([])  # menambahkan panjang array dimensi kedua
      for k in range(input.shape[2]):
        out[i][j].append([]) # menambahkan panjang array dimensi ketiga
        out[i][j][k] = euclidean(input[i][j],db[k]) # mengisi nilai aray ketiga dengan nilai dari fungsi euclidean

  return out
```

```python
# menghitung cosinus antar data daun
def cosinus(x1,x2):
  # jumlah seluruh dari x1*x2
  total_atas = sum(x1*x2) 
  # np.square untuk mengkalikan dirinya (x*x) x untuk masing masing element dari array
  total_bawah = math.sqrt(sum(np.square(x1))) * math.sqrt(sum(np.square(x2))) 
  return total_atas / total_bawah

# menghitung seluruh cosinus antar input dalam bentuk array 3d dan data daun dalam bentuk array 2d
def count_all_cosinus(input,db):
  out = []
  for i in range(input.shape[0]):
    out.append([])
    for j in range(input.shape[1]):
      out[i].append([])
      for k in range(input.shape[2]):
        out[i][j].append([])
        out[i][j][k] = cosinus(input[i][j],db[k])

  return out
```

```python
# Konversi array 3D ke 2D
def arr3dto2d(arr):
  out = []
  for y in arr:
    for x in y:
      out.append(x)
  return out


# mendapatkan nilai terendah
def getMin(arr):
  out = []
  for x in arr:
      out.append(np.argsort(x)[0])
  
  return out

# mendapatkan nilai tertinggi
def getMax(arr):
  out = []
  for x in arr:
      out.append(np.argsort(x)[3])
  
  return out

# Menghitung Akurasi
def accuracy(arr,data):
  error = 0
  value = ["A","B","C","D"] 
  for i in range(len(arr)):
   if value[arr[i]] != data[i]:
    error += 1

  return ((len(arr)-error)/len(arr))*100
```

#### functions

```python
def confusion_m(actual, predicted, label=["A", "B", "C", "D"]):
  c = confusion_matrix(actual,predicted)
  df = pd.DataFrame(data=c, columns=label, index=label)
  return df.T
```

```python
def summary_confusion_m(base, outputs, label=["A", "B", "C", "D"]):
  confusionM = []
  start = 0
  end = base
  for i in range(4):
    totalOutput = sum([outputs[j][start:end].count(i) for j in range(len(outputs))])
    confusionM.append(totalOutput)
    # confusionM.append(output1[start:end].count(i) + output2[start:end].count(i))
    start += base
    end += base

  df = pd.DataFrame({
      "DAUN":DAUN,
      "Total":confusionM
  })
  total_len_output = sum([len(out) for out in outputs])
  print('accuracy: ', (sum(confusionM)/total_len_output)*100)
  print(sum(confusionM), '/', total_len_output)

  return df
```

## menggunakan function

## function folding

```python
def folding(data, num_of_fold):
  sections = [] # tempat untuk menyimpan setiap input dan database pada suatu folding
  """
  sections = [
    #1 [input, db],
    #2 [input, db],
    ...
  ]
  """
  temp_daun = [] # tempat sementara untuk menyimpan daun2 yang sudah di-split berdasar besaran folding
  """
  temp_daun = [
    daunA [daunA[0:5], daunA[5:10, ...]],
    daunB [daunB[0:5], daunB[5:10, ...]],
    ...
  ]
  """
  for i in range(len(data)):
    temp_daun.append([])

  for index, daun in enumerate(data):
    base = int(data.shape[1]/num_of_fold)
    for i in range(0, len(daun), base):
      temp_daun[index].append(daun[i:i+base])

  # for loop diatas digunakan untuk menjadikan array temp_daun seperti yang tertera diatas.

  for i in range(num_of_fold):
    sections.append([])

  for index in range(len(sections)):
    temp_input = []
    temp_db = []
    for jenis, daun in enumerate(temp_daun):
      temp_input.append(daun[index])
      temp_db.append(np.average(np.concatenate(([ daun[i] for i in range(num_of_fold) if i != index] ), axis=0), axis=0))
    sections[index].append(temp_input)  # input
    sections[index].append(temp_db) # database

  # for loop diatas digunakan untuk menjadikan array sections seperti yang tertera diatas.

  return sections
```
