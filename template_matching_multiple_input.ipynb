# Pendefinisian Function

````python
# Konversi array 3D ke 2D
def arr3dto2d(arr):
  out = []
  for y in arr:
    for x in y:
      out.append(x)
  return out


# mendapatkan nilai terendah
def getMin(arr):
  out = []
  for x in arr:
      out.append(np.argsort(x)[0])
  
  return out

# mendapatkan nilai tertinggi
def getMax(arr):
  out = []
  for x in arr:
      out.append(np.argsort(x)[3])
  
  return out

# Menghitung Akurasi
def accuracy(arr,data):
  error = 0
  value = ["A","B","C","D"] 
  for i in range(len(arr)):
   if value[arr[i]] != data[i]:
    error += 1

  return ((len(arr)-error)/len(arr))*100
````

````python
DAUN = ["A","B","C","D"] # Untuk Menyimpan jenis Daun yang digunakan
DAUN_INPUT = ["A11","A12","A13","A14","A15", 
              "B11","B12","B13","B14","B15",
              "C11","C12","C13","C14","C15",
              "D11","D12","D13","D14","D15"] # Nama Data Daun yang digunakan sebagai input
DAUN_INPUT_2 = ["A11","A12","A13","A14","A15", "A16","A17","A18","A19","A20",
                "B11","B12","B13","B14","B15", "B16","B17","B18","B19","B20",
                "C11","C12","C13","C14","C15", "C16","C17","C18","C19","C20",
                "D11","D12","D13","D14","D15", "D16","D17","D18","D19","D20",]        
````

# EUCLIDEAN Dengan Input 5 Data

````python
# Perhitungan Euclidean
def euclidean(x1,x2):
  sum = 0
  for i in range(x1.shape[0]): 
    sum += math.pow(x1[i] - x2[i],2)
  return math.sqrt(sum)

# Menghitung seluruh Euclidean Daun A,B,C,D
def count_all_euclidean(input,db):
  out = [] # array kosong
  for i in range(input.shape[0]):
    out.append([]) # menambahkan panjang array dimensi pertama
    for j in range(input.shape[1]):
      out[i].append([])  # menambahkan panjang array dimensi kedua
      for k in range(input.shape[2]):
        out[i][j].append([]) # menambahkan panjang array dimensi ketiga
        out[i][j][k] = euclidean(input[i][j],db[k]) # mengisi nilai aray ketiga dengan nilai dari fungsi euclidean

  return out
````

## Menghitung nilai EUCLIDEAN data [A1,B1,C1,D1] dengan [A11:A15,B11:B15,C11:C15,D11:D15]

````python
KOLOM = ["Input\DB","A1","B1","C1","D1","OUTPUT"] # Sebagai Nama Kolom untuk Data Frame
data = count_all_euclidean(INPUT_1,DB) # mengambil data euclidean dari input dengan database
data2d = arr3dto2d(data) # mengubah `data` menjadi array 2 dimensi
output = getMin(data2d) # mengambil output daun
akurasi = accuracy(output,EXPECTED_VAL) # menghitung akurasi ketepatan template matching

for i in range(len(data2d)):
  data2d[i].insert(0,DAUN_INPUT[i])
  data2d[i].insert(len(data2d[i]),DAUN[output[i]])

print("Akurasi : %d"%akurasi)
df = pd.DataFrame(data2d,columns=KOLOM)
df
````

# Cosinus Simlikiti

````python
# menghitung cosinus antar data daun
def cosinus(x1,x2):
  # jumlah seluruh dari x1*x2
  total_atas = sum(x1*x2) 
  # np.square untuk mengkalikan dirinya (x*x) x untuk masing masing element dari array
  total_bawah = math.sqrt(sum(np.square(x1))) * math.sqrt(sum(np.square(x2))) 
  return total_atas / total_bawah

# menghitung seluruh cosinus antar input dalam bentuk array 3d dan data daun dalam bentuk array 2d
def count_all_cosinus(input,db):
  out = []
  for i in range(input.shape[0]):
    out.append([])
    for j in range(input.shape[1]):
      out[i].append([])
      for k in range(input.shape[2]):
        out[i][j].append([])
        out[i][j][k] = cosinus(input[i][j],db[k])

  return out
````

## Menghitung nilai COSINUS data [A1,B1,C1,D1] dengan [A11:A15,B11:B15,C11:C15,D11:D15]

````python
KOLOM = ["Input\DB","A1","B1","C1","D1","OUTPUT"] # Sebagai Nama Kolom untuk Data Frame
data = count_all_cosinus(INPUT_1,DB) # mengambil data cosinus dari input dengan database
data2d = arr3dto2d(data) # mengubah `data` menjadi array 2 dimensi
output = getMax(data2d) # mengambil output daun
akurasi = accuracy(output,EXPECTED_VAL) # menghitung akurasi ketepatan template matching

for i in range(len(data2d)):
  data2d[i].insert(0,DAUN_INPUT[i])
  data2d[i].insert(len(data2d[i]),DAUN[output[i]])

print("Akurasi : %d"%akurasi)
df = pd.DataFrame(data2d,columns=KOLOM)
df
````
